```python
import pandas as pd
import io

#Read data (has been joined/cleaned with PowerBI)
df = pd.read_csv('Time_Lapsed.csv')
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
#Remove outlier
temp = df[(df['Days_Elapsed'] >=0) & (df['Days_Elapsed'] <14)]

#Mean & STD
mu = np.mean(temp['Days_Elapsed'])
sigma = np.std(temp['Days_Elapsed'])

#Graph
_= plt.hist(temp['Days_Elapsed'],  histtype='bar', align='mid', orientation='vertical', color='red', label=None, edgecolor= 'k')
_=plt.xlabel('Days Elapsed')
_=plt.ylabel('Frequency')
_=plt.title('Days Elapsed from Start to End', fontweight='bold')
_= plt.axvline(np.mean(temp['Days_Elapsed']), color='black', linestyle='dashed', linewidth=2, label='mean')
_= plt.axvline(np.median(temp['Days_Elapsed']), color='green', linestyle='dashed', linewidth=2, label='median')
_=plt.legend()
plt.show

print('Median: ', round(np.median(temp['Days_Elapsed']), 2))
print('Mean:   ', round(np.mean(temp['Days_Elapsed']),2))
print('Sd:     ', round(np.std(temp['Days_Elapsed']),2))
```
